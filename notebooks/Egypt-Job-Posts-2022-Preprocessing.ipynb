```python
import pandas as pd
```

```python
jobsDF = pd.read_csv('../data/Egypt Job Posts 22.csv')
jobsDF
```

```python
jobsDF.info()
```

```python
jobsDF.isna().sum()
```

```python
clean_data = jobsDF.drop(['Area', 'Gender'], axis=1)
clean_data
```

```python
clean_data.isna().sum()
```

```python
clean_data = clean_data.dropna(axis=0, how='any', subset=['Company Name', 'Job Type'])
clean_data.isna().sum()
```

```python
clean_data["Governorate"] = clean_data["Governorate"].fillna('Cairo')
clean_data["Min Yrs of Excperience"] = clean_data["Min Yrs of Excperience"].fillna(0)
clean_data["Max Yrs of Experience"] = clean_data["Max Yrs of Experience"].fillna(0)
clean_data.isna().sum()
```

```python
clean_data["Skills - 1.1"] = clean_data["Skills - 1.1"].fillna('None')
clean_data["Skills - 1.2"] = clean_data["Skills - 1.2"].fillna('None')
clean_data["Skills - 1.3"] = clean_data["Skills - 1.3"].fillna('None')
clean_data["Skills - 2"] = clean_data["Skills - 2"].fillna('None')
clean_data["Skills - 2.1"] = clean_data["Skills - 2.1"].fillna('None')
clean_data["Skills - 3"] = clean_data["Skills - 3"].fillna('None')
clean_data["Skills - 3.1"] = clean_data["Skills - 3.1"].fillna('None')
clean_data["Skills - 3.2"] = clean_data["Skills - 3.2"].fillna('None')
clean_data["Skills - 3.3"] = clean_data["Skills - 3.3"].fillna('None')
clean_data["Skills - 4"] = clean_data["Skills - 4"].fillna('None')
clean_data["Skills - 5"] = clean_data["Skills - 5"].fillna('None')
clean_data["Skills - 6"] = clean_data["Skills - 6"].fillna('None')
clean_data.isna().sum()
```

```python
clean_data.duplicated().sum()
```

```python
clean_data = clean_data.drop_duplicates()
clean_data.duplicated().sum()
```

```python
clean_data.to_csv('../data/Preprocessed Egypt Job Posts 22.csv', index=False)
```
